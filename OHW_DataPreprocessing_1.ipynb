```python
df_met2008 = pd.read_csv('Met2008.csv', low_memory=False)
df_met2008
```

```python
df_met2008.dtypes
```

```python
df_met2008["time"] = pd.to_datetime(df_met2008["time"], errors="coerce")
for c in df_met2008.columns:
    if c != "time":
        df_met2008[c] = pd.to_numeric(df_met2008[c], errors="coerce")

df_met2008 = df_met2008.set_index("time").sort_index()

dep_col = None
for c in df_met2008.columns:
    if c.strip() == "Deloyment Number":
        dep_col = c
        break

df_met2008_hourly = df_met2008.resample("h").mean(numeric_only=True)

full_idx = pd.date_range(df_met2008_hourly.index.min(),
                         df_met2008_hourly.index.max(), freq="h")
df_met2008_hourly = df_met2008_hourly.reindex(full_idx)
df_met2008_hourly.index.name = "time"
```

```python
df_met2008_hourly
```

```python
if dep_col is not None:
    dep_hourly = (
        df_met2008[[dep_col]]
        .resample("h")
        .ffill()
        .reindex(full_idx)
    )
    dep_hourly[dep_col] = dep_hourly[dep_col].ffill().bfill()
    df_met2008_hourly[dep_col] = dep_hourly[dep_col]
```

```python
num_cols = [c for c in df_met2008_hourly.columns if c != dep_col]
df_met2008_hourly[num_cols] = df_met2008_hourly[num_cols].interpolate(method="time", limit=6, limit_area="inside")
```

```python
df_met2008_hourly = df_met2008_hourly.replace([np.inf, -np.inf], np.nan)
df_met2008_hourly = df_met2008_hourly.fillna(-1000)
```

```python
safe_cols = [                 
    ' Absolute Salinity Averaged Data ',
    ' Practical Salinity Averaged Data ',
    ' Temperature Averaged Data ',
    ' Potential Temperature 0 db Averaged Data ',
    ' Density Averaged Data ',
    ' Sigma t Averaged Data ',
    ' Sigma theta 0 dbar Averaged Data ',
    ' Pressure Averaged Data ',
    ' Pressure leveled Averaged Data ',
    ' Oxygen Concentration mll Averaged Data '   
]
```

```python
DEP_COL = ' Deloyment Number '
tv_cols = [c for c in safe_cols if c != DEP_COL]

horizons = [1, 3, 6, 12, 24, 48, 72]
momentum_pairs = [(1, 24), (3, 24), (6, 24)]
pct_horizons = [1, 24]

feature_blocks = []


base = df_env_all[[DEP_COL] + tv_cols].copy()
feature_blocks.append(base)


lag_feats = {}
for col in tv_cols:
    base_clean = col.strip()
    for h in horizons:
        lag_feats[f"{base_clean}_lag{h}h"] = df_env_all[col].shift(h)
feature_blocks.append(pd.DataFrame(lag_feats, index=df_env_all.index))


delta_feats = {}
for col in tv_cols:
    base_clean = col.strip()
    for h in horizons:
        lag_col = f"{base_clean}_lag{h}h"
        delta_feats[f"{base_clean}_delta{h}h"] = df_env_all[col] - feature_blocks[1][lag_col]
feature_blocks.append(pd.DataFrame(delta_feats, index=df_env_all.index))


momentum_feats = {}
for col in tv_cols:
    base_clean = col.strip()
    for h1, h2 in momentum_pairs:
        c1 = f"{base_clean}_lag{h1}h"
        c2 = f"{base_clean}_lag{h2}h"
        momentum_feats[f"{base_clean}_delta{h1}to{h2}h"] = feature_blocks[1][c1] - feature_blocks[1][c2]
feature_blocks.append(pd.DataFrame(momentum_feats, index=df_env_all.index))


pct_feats = {}
eps = 1e-12
for col in tv_cols:
    base_clean = col.strip()
    for h in pct_horizons:
        lag_col = f"{base_clean}_lag{h}h"
        lag_vals = feature_blocks[1][lag_col].values
        x_vals   = df_env_all[col].values
        denom = np.maximum(np.abs(lag_vals), eps)
        pct_feats[f"{base_clean}_pctDelta{h}h"] = (x_vals - lag_vals) / denom
feature_blocks.append(pd.DataFrame(pct_feats, index=df_env_all.index))



df_env_all = pd.concat(feature_blocks, axis=1)

df_env_all = df_env_all.iloc[max(horizons):].copy()
```

### Lets reduce the dimension of target to make this things simple and that makes sense both for data science and oceanography bioligical phenomena.

```python
import re
def parse_depth(col):
    m = re.search(r"(\d+)_m$", col)
    return float(m.group(1)) if m else None

depths = {col: parse_depth(col) for col in df_hourly.columns}

depth_cols_valid = [c for c, d in depths.items() if d is not None]
depth_cols_sorted = sorted(depth_cols_valid, key=lambda c: depths[c])
```

```python
def depth_val(c): return depths[c]

surface_cols = [c for c in depth_cols_sorted if depth_val(c) <= 20]
thermo_cols  = [c for c in depth_cols_sorted if 20 < depth_val(c) <= 40]
mid_cols     = [c for c in depth_cols_sorted if 40 < depth_val(c) <= 60]
deep_cols    = [c for c in depth_cols_sorted if 60 < depth_val(c) <= 85]
```

```python
df_targets = pd.DataFrame(index=df_hourly.index)
df_targets["surface_mean"]     = df_hourly[surface_cols].mean(axis=1)
df_targets["thermocline_mean"] = df_hourly[thermo_cols].mean(axis=1)
df_targets["mid_mean"]         = df_hourly[mid_cols].mean(axis=1)
df_targets["deep_mean"]        = df_hourly[deep_cols].mean(axis=1)

df_targets["delta_surface_thermo"] = df_targets["surface_mean"]     - df_targets["thermocline_mean"]
df_targets["delta_thermo_mid"]     = df_targets["thermocline_mean"] - df_targets["mid_mean"]
df_targets["delta_mid_deep"]       = df_targets["mid_mean"]         - df_targets["deep_mean"]
```

```python
df_deltas = df_hourly[depth_cols_sorted].diff(axis=1).iloc[:, 1:]
df_deltas.columns = [
    f"delta_{depth_cols_sorted[i-1]}_{depth_cols_sorted[i]}"
    for i in range(1, len(depth_cols_sorted))
]
```

```python
df_deltas = df_hourly[depth_cols_sorted].diff(axis=1).iloc[:, 1:]
df_deltas.columns = [
    f"delta_{depth_cols_sorted[i-1]}_{depth_cols_sorted[i]}"
    for i in range(1, len(depth_cols_sorted))
]
```

```python
var = df_deltas.var()
q = 0.30  # keep ~80% most variable; adjust (e.g., 0.3 to keep 70%)
keep = var[var > var.quantile(q)].index
df_deltas = df_deltas[keep]
```

```python
corr = df_deltas.corr().abs()
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
to_drop = [c for c in upper.columns if (upper[c] > 0.90).any()]
df_deltas = df_deltas.drop(columns=to_drop)
```

```python
cols_means_and_layer_deltas = [
    "surface_mean","thermocline_mean","mid_mean","deep_mean",
    "delta_surface_thermo","delta_thermo_mid","delta_mid_deep"
]
```

```python
df_targets_reduced = pd.concat([df_targets[cols_means_and_layer_deltas], df_deltas], axis=1)
```
